### Nationwide
This report lists all of the enforcement cases initiated by PHMSA in a given year, beginning in 2002. The yearly tables include the date the case was opened, the name of the operator involved, the PHMSA Region initiating the enforcement action(s), the corresponding date on which the case was closed (if applicable), the current status of the case, and PHMSA's unique case number. Separate reports provide a listing of cases based on the type of enforcement action involved.
https://primis.phmsa.dot.gov/comm/reports/enforce/CasesOpen_opid_0.html?nocache=5321


```python
cases_initiated_2002_to_2017 = pd.read_html('https://primis.phmsa.dot.gov/comm/reports/enforce/CasesOpen_opid_0.html?nocache=5321', header=0)
len(cases_initiated_2002_to_2017)
```

```python
df_2017 = cases_initiated_2002_to_2017[0]
df_2017.head()
```

```python
cases = df_2017
for df in cases_initiated_2002_to_2017[1:]:
    cases = pd.concat([cases, df], ignore_index=True)
```
